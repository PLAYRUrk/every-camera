```python
def getOnOffAngles(t_date,loc,ang):
    not_sunrise=True
    not_sunset=True
    for h in range(24):
        for m in range(60):
                
            sun_time = t_date+timedelta(hours=h,minutes=m)
            '''
            obs.date = date
            sun = ephem.Sun(obs)
            sun.compute(obs)
            '''
            altaz = AltAz(obstime=Time(sun_time, scale='utc'), location=loc)
            #print(str(h)+':'+str(m),np.degrees(sun.alt))
            sun_alt = get_sun(Time(sun_time, scale='utc')).transform_to(altaz).alt.degree
            if h!=0 and m!=0:
                
                if sun_alt>prev_ang:
                    sunrise=True
                else:
                    sunrise=False
                #print(np.degrees(sun.alt),sunrise)
                if sunrise and not_sunrise and sun_alt>ang and sun_alt<ang+1:
                    sunrise_time=sun_time
                    sr_ang=sun_alt
                    not_sunrise=False
                    #print('sunrise_time',sunrise_time,sr_ang)
                if not sunrise and not_sunset and sun_alt<ang:
                    sunset_time=sun_time
                    ss_ang=sun_alt
                    not_sunset=False
            prev_ang=sun_alt
    if not_sunrise:
        sunrise_time=sun_time
        sr_ang=sun_alt
    if not_sunset:
        return [datetime.strftime(sun_time,'%Y-%m-%d')]
    else:
        '''
        return [datetime.strftime(date,'%Y-%m-%d'),
                datetime.strftime(sunset_time,'%H:%M'),datetime.strftime(sunrise_time,'%H:%M'),
                int(datetime.strftime(date,'%Y')),int(datetime.strftime(date,'%m')),int(datetime.strftime(date,'%d')),
                int(datetime.strftime(sunset_time,'%H')),int(datetime.strftime(sunset_time,'%M')),
                int(datetime.strftime(sunrise_time,'%H')),int(datetime.strftime(sunrise_time,'%M')),
                ss_ang,sr_ang]
        '''
        return [datetime.strftime(sun_time,'%d.%m.%Y'),
                datetime.strftime(sunset_time,'%H:%M'),datetime.strftime(sunrise_time,'%H:%M')]

'''
obs=ephem.Observer()
obs.lat='53:19:27' #Uzur
obs.lon='107:44:24' #Uzur

obs.elevation=473 #Uzur
#obs.elevation=500 #Irkutsk
#obs.elevation=39 #Istok

print(obs)
'''

loc = EarthLocation(lon=GPHO_TK.LONGITUDE, lat=GPHO_TK.LATITUDE, height=GPHO_TK.ALTITUDE)

ang=-10
beg_date=datetime(2026,1,1,0,0,0)
num_days=100
all_dates=[]

for c_date in range(num_days):
    #t = Time(dt_object, scale='utc')
    #t_date=Time(beg_date+timedelta(days=c_date), scale='utc')
    t_date=beg_date+timedelta(days=c_date)
    all_dates.append(getOnOffAngles(t_date,loc,ang))

```
